##### Getting Categories from different columns

```python
numerical_features= [features for features in df.columns
                     if df[features].dtype != 'object']
categorical_features= [features for features in df.columns
                       if df[features].dtype == 'object']

print('We have {} numberical features: {}'.format(len(numerical_features), (numerical_features)))
print('We have {} categorical features: {}'.format(len(categorical_features), (categorical_features)))
```

## Adding the scores column

```python
df['total_score']= df['math_score'] + df['reading_score'] + df['writing_score']
```

```python
df
```

## Adding average column

```python
df['average_score']= df['total_score']/3
```

```python
df
```

## Rounding the decimal numbers to 2

```python
df['average_score']= df['average_score'].round(2)
```

```python
df
```

## Checking how many students scored full marks

```python
math_full= df[df['math_score'] == 100]['math_score'].count()
reading_full= df[df['reading_score'] == 100]['reading_score'].count()
writing_full= df[df['writing_score'] == 100]['writing_score'].count()

print(f'Number of students scored 100 in Math: {math_full}')
print(f'Number of students scored 100 in Reading: {reading_full}')
print(f'Number of students scored 100 in Writing: {writing_full}')
```

## Checking who scored below 20

```python
math_20= df[df['math_score'] <= 20]['math_score'].count()
reading_20= df[df['reading_score'] <= 20]['reading_score'].count()
writing_20= df[df['writing_score'] <= 20]['writing_score'].count()

print(f'Number of students scored below 20 in Math: {math_20}')
print(f'Number of students scored below 20 in Math: {reading_20}')
print(f'Number of students scored below 20 in Math: {writing_20}')
```

##### Based on the analysis, it appears that students are excelling in reading but facing challenges in math.

## Exploring the data through visualization

```python
plt.subplots(1,2, figsize= (14,7))
plt.subplot(121)
sns.histplot(data= df, x= 'average_score', kde= True, color= 'b')
plt.title('Average Score')
plt.grid()
plt.subplot(122)
sns.histplot(data=df, x= 'average_score', kde= True, hue= 'gender')
plt.title('Average Score based on Gender')
plt.grid()
plt.show()
```

##### In the above graph
>As per the above obsevation we see that female students have scored the highest.

```python
plt.subplots(1,3, figsize= (20,6))
plt.subplot(131)
sns.histplot(data= df, x= 'average_score', kde= True, hue= 'parental_level_of_education')
plt.title('Average score based on Parental Education')
plt.grid()
plt.subplot(132)
sns.histplot(data= df[df.gender=='female'], x= 'average_score', kde= True, hue= 'parental_level_of_education')
plt.title('Average score based on Parental Education on female')
plt.grid()
plt.subplot(133)
sns.histplot(data= df[df.gender=='male'], x= 'average_score', kde= True, hue= 'parental_level_of_education')
plt.title('Average score based on Parental Education on male')
plt.grid()
```

##### In the above graphs
> Students whose parents have an Associate’s degree tend to score more than 60 marks.

> In the second graph, female students perform well when their parents have a degree from ‘some college.’

> Male students tend to perform well in exams when their parents have an Associate’s degree.
